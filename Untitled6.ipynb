```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Generar el DataFrame con los datos de ejemplo
data = []
for _ in range(15000):
    complejidad = np.random.choice(['Baja', 'Media', 'Alta'], p=[0.4, 0.4, 0.2])
    horas_estimadas = np.random.randint(2, 25)

    # Aadir sesgo basado en complejidad
    if complejidad == 'Alta':
        factor = np.random.choice([0.8, 1.0, 1.5, 2.0], p=[0.2, 0.5, 0.2, 0.1])
    else:
        factor = np.random.uniform(0.7, 1.3)

    data.append({
        'Tarea_ID': str(np.random.randint(1000, 9999)),
        'Equipo': np.random.choice(['Frontend', 'Backend', 'QA']),
        'Complejidad': complejidad,
        'Horas_Estimadas': horas_estimadas,
        'Horas_Reales': round(horas_estimadas * factor, 2),
        'Responsable': np.random.choice(['Casey Moore', 'James Owens', 'Susan Reynolds', 'Tracy Moore', 'Alyssa Marshall'])
    })

df_desarrollo = pd.DataFrame(data)

# Clculo del porcentaje de error por tarea
df_desarrollo['Error'] = abs(df_desarrollo['Horas_Reales'] - df_desarrollo['Horas_Estimadas']) / df_desarrollo['Horas_Estimadas'] * 100

# 1. Anlisis por Complejidad (porcentaje de error promedio)
error_por_complejidad = df_desarrollo.groupby('Complejidad')['Error'].mean().reset_index()

# Visualizacin del error por Complejidad
plt.figure(figsize=(8, 6))
sns.barplot(data=error_por_complejidad, x='Complejidad', y='Error', palette='viridis')
plt.title("Error de estimacin por Complejidad")
plt.xlabel("Complejidad")
plt.ylabel("Porcentaje de Error")
plt.show()

# 2. Identificacin de equipos con mayor desviacin
desviacion_por_equipo = df_desarrollo.groupby('Equipo')['Error'].mean().reset_index()

# Visualizacin de la desviacin por Equipo
plt.figure(figsize=(8, 6))
sns.barplot(data=desviacion_por_equipo, x='Equipo', y='Error', palette='coolwarm')
plt.title("Desviacin media por Equipo")
plt.xlabel("Equipo")
plt.ylabel("Porcentaje de Error")
plt.show()

# 3. Ajustes con percentiles histricos
percentil_90 = df_desarrollo.groupby('Complejidad')['Horas_Reales'].quantile(0.90).reset_index()
percentil_10 = df_desarrollo.groupby('Complejidad')['Horas_Reales'].quantile(0.10).reset_index()

# Visualizacin de los percentiles
plt.figure(figsize=(8, 6))
sns.lineplot(data=percentil_90, x='Complejidad', y='Horas_Reales', label="Percentil 90", marker='o')
sns.lineplot(data=percentil_10, x='Complejidad', y='Horas_Reales', label="Percentil 10", marker='o')
plt.title("Percentiles histricos por Complejidad")
plt.xlabel("Complejidad")
plt.ylabel("Horas Reales")
plt.legend()
plt.show()

# 4. Visualizacin de los efectos de la poblacin y muestras
plt.figure(figsize=(8, 6))
sns.boxplot(data=df_desarrollo, x='Complejidad', y='Error', palette='Set3')
plt.title("Distribucin del error por Complejidad")
plt.xlabel("Complejidad")
plt.ylabel("Error (%)")
plt.show()
```
